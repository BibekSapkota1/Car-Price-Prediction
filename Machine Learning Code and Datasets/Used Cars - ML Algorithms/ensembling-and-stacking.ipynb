```python
import numpy as np
import pandas as pd

# Assuming your DataFrame is called 'df'
def exponential_mean_log1p(predictions):
    # Apply log1p transformation
    log_preds = np.log1p(predictions)
    
    # Calculate mean of log-transformed predictions
    mean_log_preds = np.mean(log_preds, axis=1)
    
    # Apply exponential (inverse of log1p)
    final_predictions = np.expm1(mean_log_preds)
    
    return final_predictions


# Select prediction columns
pred_columns = ['xgb_pred', 'lgb_pred', 'cat_pred', 'aada_pred', 'decision_pred','random_pred',	'linear_pred','lasso_pred','ridge_pred','pls_pred',	'Neural_pred' ]

# Calculate final predictions
combined_df['final_pred'] = exponential_mean_log1p(combined_df[pred_columns])

# Display the results
combined_df.head()
```

```python
import numpy as np
import pandas as pd
import optuna
from sklearn.model_selection import train_test_split

def rmsle(y_true, y_pred):
    """
    Root Symmetric Mean Log Error
    Handles cases with zero and negative values
    """
    # Clip values to prevent log issues
    y_true = np.clip(y_true, 1e-6, None)
    y_pred = np.clip(y_pred, 1e-6, None)
    
    # Calculate log error
    log_error = np.log1p(y_true) - np.log1p(y_pred)
    
    # Calculate RMSLE
    return np.sqrt(np.mean(log_error**2))

def objective(trial):
    # Prepare the data
    X = df[['OOF_Pred_XGBoost', 'OOF_Pred_LightGBM', 'OOF_Pred_CatBoost',
            'OOF_Pred_AdaBoost', 'OOF_Pred_DecisionTree', 'OOF_Pred_RandomForest',
            'OOF_Pred_LinearRegression', 'OOF_Pred_Lasso', 'OOF_Pred_Ridge',
            'OOF_Pred_PLS', 'OOF_Pred_NN']]
    y = df['Actual']
    
    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # Optimize 11 weights (one per model)
    weights = np.array([trial.suggest_float(f'w{i}', 0, 1) for i in range(11)])
    
    # Normalize the weights
    total_weight = np.sum(weights)
    normalized_weights = weights / total_weight

    # Calculate weighted prediction
    pred = np.average(X_test, axis=1, weights=normalized_weights)
    
    # Calculate RMSLE
    rmsle_score = rmsle(y_test, pred)
    
    return rmsle_score

# Ensure df is properly defined
df = combined_oof_df  # Ensure `combined_oof_df` is loaded in your environment

# Create a study object and optimize
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=500)

# Print results
print("Best trial:")
trial = study.best_trial
print("  Best RMSLE: ", trial.value)
print("  Best Weights:")
best_weights = np.array([trial.params[f'w{i}'] for i in range(11)])

# Normalize final weights
total_weight = np.sum(best_weights)
normalized_weights = best_weights / total_weight

for i, model in enumerate(['XGBoost', 'LightGBM', 'CatBoost', 'AdaBoost', 'DecisionTree', 
                           'RandomForest', 'LinearRegression', 'Lasso', 'Ridge', 'PLS', 'NN']):
    print(f"    {model}: {normalized_weights[i]}")

# Final weighted prediction using best weights
final_pred = np.average(df[['OOF_Pred_XGBoost', 'OOF_Pred_LightGBM', 'OOF_Pred_CatBoost',
                            'OOF_Pred_AdaBoost', 'OOF_Pred_DecisionTree', 'OOF_Pred_RandomForest',
                            'OOF_Pred_LinearRegression', 'OOF_Pred_Lasso', 'OOF_Pred_Ridge',
                            'OOF_Pred_PLS', 'OOF_Pred_NN']], 
                        axis=1, 
                        weights=normalized_weights)

df['Weighted_Prediction'] = final_pred

# Display final results
print("\nFinal DataFrame with Weighted Prediction:")
print(df)
```

```python
import optuna
import xgboost as xgb
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import mean_squared_log_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split

# Define RMSLE metric
def rmsle(y_true, y_pred):
    return np.sqrt(mean_squared_log_error(y_true, y_pred))


# Define numerical and categorical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.drop(['Actual', 'Weighted_Prediction', 'ID']).tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Convert categorical columns to category type
for col in categorical_cols:
    df[col] = df[col].astype('category')

# Define preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'
)

# Prepare train and validation sets


train_df, valid_df = train_test_split(df, test_size=0.2, random_state=42)

y_train = train_df['Actual']
y_valid = valid_df['Actual']

# Drop unnecessary columns before transformation
X_train = train_df.drop(['Actual', 'Weighted_Prediction', 'ID'], axis=1)
X_valid = valid_df.drop(['Actual', 'Weighted_Prediction', 'ID'], axis=1)

# Apply preprocessor
X_train = preprocessor.fit_transform(X_train)
X_valid = preprocessor.transform(X_valid)

# Define objective function for Optuna
def objective(trial):
    # Define hyperparameter search space
    params = {
        'objective': 'reg:squarederror',
        'tree_method': 'hist',
        'device': 'cuda',
        'eval_metric': 'rmsle',
        'seed': 42,
        'max_delta_step': 1,
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_child_weight': trial.suggest_float('min_child_weight', 1, 10),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'lambda': trial.suggest_float('lambda', 1e-3, 10.0, log=True),
        'alpha': trial.suggest_float('alpha', 1e-3, 10.0, log=True)
    }

    # Create DMatrix for XGBoost
    dtrain = xgb.DMatrix(data=X_train, label=y_train)
    dvalid = xgb.DMatrix(data=X_valid, label=y_valid)

    # Train the model
    model = xgb.train(
        params=params,
        dtrain=dtrain,
        num_boost_round=200,
        evals=[(dvalid, 'valid')],
        early_stopping_rounds=50,
        verbose_eval=0
    )

    # Predict on validation set
    val_preds = model.predict(dvalid)
    return rmsle(y_valid.values, val_preds)

# Optimize hyperparameters using Optuna
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=50)

# Print best trial parameters
print(f"Best trial parameters: {study.best_params}")
```

```python
import optuna
import xgboost as xgb
import numpy as np
import pandas as pd
import pickle
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_log_error

# Define RMSLE metric
def rmsle(y_true, y_pred):
    return np.sqrt(mean_squared_log_error(y_true, y_pred))


# Ensure column names are stripped of extra spaces
train.columns = train.columns.str.strip()
test.columns = test.columns.str.strip()

# Check if 'kfold' exists; if not, create it
if 'kfold' not in train.columns:
    folds = 5
    train["kfold"] = -1
    kf = KFold(n_splits=folds, shuffle=True, random_state=42)
    
    for fold, (train_idx, valid_idx) in enumerate(kf.split(train)):
        train.loc[valid_idx, "kfold"] = fold

# Drop 'Actual' and 'kfold' safely
train_cols_to_drop = [col for col in ["Actual", "kfold"] if col in train.columns]
X_train = train.drop(columns=train_cols_to_drop, errors="ignore")
y_train = train["Actual"] if "Actual" in train.columns else None

# Test dataset should NOT contain 'Actual' or 'kfold'
X_test = test.copy()

# Drop 'ID' safely
if "ID" in X_train.columns:
    X_train = X_train.drop(columns=["ID"])
if "ID" in X_test.columns:
    X_test = X_test.drop(columns=["ID"])


# Define numerical and categorical columns
categorical_cols = X_train.select_dtypes(include=['category', 'object']).columns.tolist()
numerical_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Define preprocessor for encoding and scaling
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'
)

folds = 5  # Number of folds
xgb_models = []
val_ll = []
test_predictions_xgb = np.zeros(len(test))

for fold in range(folds):
    print(f"Training fold {fold + 1}")

    # Split train into training and validation sets
    xtrain = train[train.kfold != fold].reset_index(drop=True)
    xvalid = train[train.kfold == fold].reset_index(drop=True)

    # Ensure validation set is correctly handled
    ytrain = xtrain["Actual"]
    yvalid = xvalid["Actual"]

    # Drop unnecessary columns safely
    xtrain = xtrain.drop(columns=["Actual", "ID", "kfold"], errors="ignore")
    xvalid = xvalid.drop(columns=["Actual", "ID", "kfold"], errors="ignore")
    X_test_cleaned = X_test.drop(columns=["ID"], errors="ignore")

    # Apply preprocessing
    xtrain_preprocessed = preprocessor.fit_transform(xtrain)
    xvalid_preprocessed = preprocessor.transform(xvalid)
    X_test_preprocessed = preprocessor.transform(X_test_cleaned)

    # Create DMatrix for train, validation, and test
    dtrain = xgb.DMatrix(data=xtrain_preprocessed, label=ytrain)
    dvalid = xgb.DMatrix(data=xvalid_preprocessed, label=yvalid)
    dtest = xgb.DMatrix(data=X_test_preprocessed)

    # Define XGBoost parameters
    params = {
        'learning_rate': 0.2998728525343831,
        'max_depth': 5,
        'min_child_weight': 6.452118656451711,
        'subsample': 0.5214672023025306,
        'colsample_bytree': 0.7735782565442656,
        'lambda': 0.028233433857777393,
        'alpha': 2.1531425856868864,
        'objective': 'reg:squarederror',
        'tree_method': 'hist',  # Use GPU if available
        'device': 'cuda',
        'eval_metric': 'rmsle',
        'max_delta_step': 1,
        'seed': 42,
    }

    # Train the model
    model = xgb.train(
        params=params,
        dtrain=dtrain,
        num_boost_round=10000,
        evals=[(dvalid, 'valid')],
        early_stopping_rounds=300,
        verbose_eval=0
    )

    # Predictions for validation set
    val_preds = model.predict(dvalid)

    # Calculate RMSLE
    val_score = rmsle(yvalid.values, val_preds)
    print(f"Fold {fold} RMSLE: {val_score:.5f}")
    print("*" * 50)

    # Store validation score and model
    val_ll.append(val_score)
    xgb_models.append(model)

    # Predictions for test set (averaged over folds)
    test_preds = model.predict(dtest) / folds
    test_predictions_xgb += test_preds

# Save the trained stacking model as a pickle file
with open("xgb_stacking_model.pkl", "wb") as f:
    pickle.dump(xgb_models, f)

# Print mean validation RMSLE
print(f"Mean Root Mean Square Log Loss: {np.mean(val_ll)}")

# Save test predictions to CSV
test["XGB_Stacking_Prediction"] = test_predictions_xgb
test[["ID", "XGB_Stacking_Prediction"]].to_csv("stacking_predictions.csv", index=False)

print("\nStacking model and predictions saved successfully!")
```
